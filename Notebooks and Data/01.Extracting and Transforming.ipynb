```python
import pandas as pd
```

```python
data=pd.read_csv("rideshare_kaggle.csv")
```

```python
data = data.drop(columns='id')
```

```python
data.isna().sum()
```

```python
data[data['price'].isna()].head()
```

```python
data['price_missing'] = data['price'].isnull()
```

```python
median_price = data['price'].median()
data['price_filled'] = data['price'].fillna(median_price)
```

```python
data.to_csv("rides_staging.csv", index=False)
data[data['price'].notnull()].to_csv("rides_cleaned.csv", index=False)
```
